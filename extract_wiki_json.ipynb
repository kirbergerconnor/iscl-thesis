```python
import json

def extract_json(json_bytes):
    try:
        data = json.loads(json_bytes.decode('utf-8'))
        text = data.get('text', None)

        return text  
    except (UnicodeDecodeError, json.JSONDecodeError):
        return None 

output = []

with open('/Users/k/Docs/School/Tuebingen/Thesis/iscl-thesis/text/AA/wiki_01', 'rb') as bf:
    content = bf.read()
    json_objects = content.split(b'\n')
    for json_str in json_objects:
        text = extract_json(json_str)
        output.append(text)
```
